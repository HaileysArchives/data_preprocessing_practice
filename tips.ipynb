```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from collections import Counter
```

```python
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv'
tips = pd.read_csv(url)

tips.head()
```

```python
tips.info()
```

```python
# Delete the Unnamed 0 column
del tips['Unnamed: 0']
```

# **sns.displot()**

```python
# Plot the total_bill column histogram

# displot vs histplot
# distplot: 히스토그램과 KDE 곡선과 함께 표시 
ttbill = sns.distplot(tips['total_bill'])

# xlabel, ylabel, title은 Axes 객체의 속성이 아니라, 메서드로 존재한다. 
ttbill.set_xlabel('Value')
ttbill.set_ylabel('Frequency')
ttbill.set_title('Total Bill')
```

# **3.1 Pyplot 방식**
### - plt.xlabel, plt.ylabel, plt.title은 Matplotlib의 현재 활성화된 Axes에 영향을 미칩니다.
### - Pyplot은 사용이 간단하지만, 여러 개의 플롯이나 복잡한 설정에서는 제어가 어렵습니다.
# **3.2 Axes 방식**
### - 객체 지향 방식에서는 Axes 객체마다 독립적으로 설정을 관리합니다.
### - 여러 축을 다룰 때 더 유연하고 직관적

# **sns.jointplot()**

```python
# Create a scatter plot presenting the relationship between total_bill and tip
sns.jointplot(x = 'total_bill', y = 'tip', data = tips)
```

# **Seaborn에서 사용할 수 있는 주요 그래프 종류**
### (1) 분포 시각화
- histoplot: 히스토그램과 KDE 곡선 시각화
- distplot: 데이터의 분포를 히스토그램이나 KDE로 표현
### (2) 관계 시각화 
- scatterplot: **두 변수 간**의 산점도
- lineplot: 연속 데이터의 추세선
### (3) **범주형** 데이터 시각화 
- barplot: 범주별 평균 값을 막대 그래프로 표시
- countplot: 범주형 데이터의 빈도수
- boxplot: 범주형 데이터의 사분위수
- **stripplot**: 범주형 데이터의 산점도
### (4) 행렬 플롯
- heatmap: 2D 데이터나 상관행렬을 히트맵으로 시각화
### (5) 다변량 시각화 
- pairplot: **여러 변수 간의 관계**를 산점도와 히스토그램으로 표현
- jointplot: **두 변수 간의 관계**와 분포를 결합하여 표현

# **Pyplot에서 사용할 수 있는 주요 그래프 종류**
### (1) 기본 플롯
- plot: 기본 선 그래프와 점 그래프
- scatter: 산점도
- bar: 막대 그래프
- barh: 가로 막대 그래프
- hist: 히스토그램
- pie: 원형 차트
### (2) 기타 그래프
- boxplot
- heatmap: 직접 행렬 데이터로 구현 필요

```python
# Create one image with the relationship of total_bill, tip and size. 
sns.pairplot(tips)
```

```python
# Present the relationship between days and total_bill value
# days가 범주형 데이터이기 때문에 
sns.stripplot(x = 'day', y = 'total_bill', data=tips, jitter=True)

# jitter(데이터 포인트 간의 중첩을 줄이는 기능) 
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data=tips, jitter=False)
```

```python
# Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex
sns.stripplot(x='tip', y='day', hue='sex', data=tips, jitter=True)
```

```python
# Create a box plot presenting the total_bill per day different the time (Dinner or Lunch)
sns.boxplot(x='day', y='total_bill', hue='time', data=tips)
```

# **sns.FacetGrid()**
### **FacetGrid.map()**
### 그래프 자체가 아닌 그래프를 그리는 틀(그리드 구조)이다. 
### 데이터셋을 조건에 따라 행과 열로 나눠 플롯을 배치하는 데 사용

```python
# Create two histograms of the tip value based for Dinner and Lucnh. They must be side by side.

df = sns.FacetGrid(tips, col = 'time')
df.map(plt.hist, 'tip')
```

# **add_legend(): 범례**

```python
# Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
# # 흡연자 또는 비흡연자에 따라 다른 총_청구서 값과 팁 관계를 나타내는 두 개의 산점도 그래프(남성용 및 여성용)를 만듭니다.

df_gender = sns.FacetGrid(tips, col = 'sex', hue = 'smoker')
df_gender.map(plt.scatter, "total_bill", "tip") # scatterplot의 x와 y에 전달 
# FacetGrid.map(): 각각의 서브플롯에 동일한 그래프를 반복적으로 그리는 역할

df_gender.add_legend() # 범례
```
