```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep="\t")
df.head(5)
```

```python
df.info()
```

df.item_price에 저장된 가격 데이터를 처리하여 **문자열에서 숫자(float)**로 변환

- 즉, $(앞부분)과 마지막 문자(일부 포맷에서는 ) 같은 기호가 있을 수도 있음)를 제외한 값을 추출
> 그래서 value[1:-1] 1부터 시작

```python
prices = [float(value[1:-1]) for value in df.item_price]

df['item_price'] = prices
```

```python
df.head(5)
```

drop_duplicates()는 Pandas 라이브러리에서 제공하는 메서드로, DataFrame이나 Series에서 중복된 행을 제거하는 데 사용
> 데이터 분석 과정에서 중복된 데이터를 정리하고, 고유한 데이터만 남기고 싶을 때 유용

```python
df_filtered = df.drop_duplicates(['item_name', 'quantity', 'choice_description'])

df_one_prod = df_filtered[df_filtered['quantity'] == 1]
df_one_prod
```

How many products cost more than $10.00?

```python
# 3. 수량(quantity)이 1인 데이터 필터링
df_one_prod = df[df['quantity'] == 1]
df_one_prod.head(5)
```

```python
# 4. $10.00 이상의 제품 필터링 및 고유 제품 수 계산
item_per_price_10 = df_one_prod[df_one_prod['item_price'] > 10]['item_name']
item_per_price_10.unique()

# 아래와 같이 배열 형태로 반환되기 때문에 count() 메서드를 직접 사용할 수 없다.
```

```python
len(item_per_price_10.unique())
```

What is the price of each item?

```python
item_per_price = df.loc[df['quantity'] == 1, :]
item_per_price = item_per_price.drop(columns=['choice_description', 'quantity', 'order_id'])
item_per_price.sort_values(by='item_price', ascending=False)
```

```python
df_filtered = df.drop_duplicates(subset=['item_name', 'quantity'])

one_prod = df_filtered.loc[df_filtered['quantity'] == 1]
per_price = one_prod[['item_name', 'item_price']]
per_price.count()
```

> 코드 1은 중복된 데이터를 제거한 뒤 작업을 수행하므로 중복을 제거한 고유 제품 기준으로 결과를 산출합니다.
- choice_description에 따라 똑같은 메뉴에 따라서도 가격이 다름


> 코드 2는 중복된 데이터가 포함된 상태에서 작업을 수행하므로 조건을 만족하는 모든 주문의 고유 제품 이름을 기반으로 결과를 계산합니다.

```python
# 중복 제거 전후 데이터 크기 비교
print("Original data size:", len(df))
print("After dropping duplicates:", len(df_filtered))

# 중복 제거 전후 조건을 만족하는 데이터 비교
print("Quantity == 1 (before dropping duplicates):", len(df[df['quantity'] == 1]))
print("Quantity == 1 (after dropping duplicates):", len(df_filtered[df_filtered['quantity'] == 1]))
```
